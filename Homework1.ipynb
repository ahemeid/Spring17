```python
from sklearn import preprocessing
import numpy as np
scaler = preprocessing.StandardScaler().fit(Data['L-CORE'])
Data['L-CORE'] = scaler.transform(Data['L-CORE'])

scaler = preprocessing.StandardScaler().fit(Data['L-SURF'])
Data['L-SURF'] = scaler.transform(Data['L-SURF'])

scaler = preprocessing.StandardScaler().fit(Data[' L-O2'])
Data[' L-O2'] = scaler.transform(Data[' L-O2'])

scaler = preprocessing.StandardScaler().fit(Data['L-BP'])
Data['L-BP'] = scaler.transform(Data['L-BP'])

scaler = preprocessing.StandardScaler().fit(Data['SURF-STBL'])
Data['SURF-STBL'] = scaler.transform(Data['SURF-STBL'])

scaler = preprocessing.StandardScaler().fit(Data[' CORE-STBL'])
Data[' CORE-STBL'] = scaler.transform(Data[' CORE-STBL'])

scaler = preprocessing.StandardScaler().fit(Data['BP-STBL'])
Data['BP-STBL'] = scaler.transform(Data['BP-STBL'])

print(Data)
```

Class labels should be converted as integers to avoid technical glitches. Furthermore, it is important for not only comparing measurements that have different units, but it's a requirements for many machine learning algorithms. Additionally, if any value has big variance, it will skew the results and hinder training the model in the instance of machine learning due to the effect of the variance, and that's why we normalized the data.
